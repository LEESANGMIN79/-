```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3,3)

grid = plt.GridSpec(3,3, wspace=0.4, hspace=0.3)
X = np.random.uniform(-2,4,100)
Y = np.random.uniform(-2,2,100)
plt.subplot(grid[0, :]).scatter(X,Y)


X = np.linspace(1.0, 4.0, 100)
Y = X
plt.subplot(grid[1, :2]).plot(X,Y,linestyle='--')


X = np.linspace(-0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[2, 0]).plot(X,Y,color='g')


X = np.arange(10)
Y = np.abs(np.random.random(10)*10)
plt.subplot(grid[2, 1]).bar(X,Y)


Z = np.random.uniform(0,1,(10,5))
plt.subplot(grid[1:, 2]).imshow(Z)
```
