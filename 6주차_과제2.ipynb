```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
weather = pd.read_csv(weather_file, encoding='CP949')
weather['month'] = pd.DatetimeIndex(weather['일시']).month
weather['day'] = pd.DatetimeIndex(weather['일시']).day
weather['hour'] = pd.DatetimeIndex(weather['일시']).hour

monthly = [ None for x in range(12) ]
monthly_wind_speed = [ 0 for x in range(12) ]
for i in range(12):
    monthly[i] = weather[weather['month'] == i+1]
    monthly_wind_speed[i] = monthly[i]['평균풍속'].mean()

months = np.arange(1,13)
plt.bar(months, monthly_wind_speed, color='green')
plt.xlabel('month')
plt.ylabel('Average wind speed')
plt.show()
```
